### "Create a bar chart or histogram  to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population."

```python
df = pd.read_csv('customer_shopping_data.csv')
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_num = df.select_dtypes(include = 'number')
df_num.head()
```

```python
df_cate = df.select_dtypes(exclude = 'number')
df_cate.head()
```

```python
df.isnull().sum()
```

there are no null values in the data set

```python
# removing invoice_no bcz it does not make any sense

df = df.drop(['invoice_no'],axis = 1)
df
```

```python
df.head()
```

```python
for i in df_cate:
    print(df[i].value_counts(normalize=True))
    print('Category:',i)
    print('****')
```

```python
df.describe()
```

```python
df = pd.DataFrame(df)

# Count the occurrences of each gender
gender_counts = df['gender'].value_counts()

# Plotting the bar chart
bars = plt.bar(gender_counts.index, gender_counts.values)

total = sum(gender_counts)
for bar in bars:
# bar.get_height() is used to retrieve the height of each bar in the bar chart. 
    height = bar.get_height()
    percentage = (height / total) * 100
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{percentage:.1f}%', ha='center', va='bottom
             
# Adding title and labels
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.show()
```

 * The dataset contains a higher proportion of males compared to females. This suggests that males are more represented in the dataset than females.
 * Knowing the gender distribution can help businesses divide their market into different groups. 
 * They can then customize their products or services based on what men and women like or how they behave differently.
